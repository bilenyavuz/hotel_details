```python
import os
import json
import requests
```

```python
hotels_dir = './hotels'
hotel_ids = [hotel_id for hotel_id in os.listdir(hotels_dir) if os.path.isdir(os.path.join(hotels_dir, hotel_id))]
print(hotel_ids)
```

```python
for hotel_id in hotel_ids:
    print(f"Downloading images for hotel {hotel_id}")
    hotel_dir = os.path.join(hotels_dir, hotel_id)
    details_file = os.path.join(hotel_dir, 'details.json')
    with open(details_file, 'r') as f:
        details = json.load(f)
        for image in details["Images"]:
            original_url = image["OriginalUrl"]
            print(f"Downloading image {original_url}")
            response = requests.get(original_url)
            if response.status_code == 200:
                image_file = os.path.join(hotel_dir, os.path.basename(original_url))
                with open(image_file, 'wb') as img:
                    img.write(response.content)
            else:
                print(f"Failed to download image {original_url}")
      
```
